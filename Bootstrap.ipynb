## Bootstraping 

Assume that we have a sample of $n$ data points generated by an **unknown** distribution. Suppose we want to estimate the mean. We can't generate new samples to compute the mean and observe the distrubution of those means. 

What can we do?

We can *bootstrap* new samples by choosing $n$ points with replacement from our data and compute the mean of each sample. 

We are sampling with replacement!

- Bootstrap sample: A sample taken with replacement from an observed dataset 
- Resampling: The process of taking repeated samples from observed data; includes both bootstrap and permutation (shuffling) procedures


```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
# parameters for Binomial Trials
p =.7
n_trials = 100
real_var = n_trials * p * (1-p)
```

```python
# Simulates n samples with n_trails of a Binomial Distribution of probability p
def simulate(n_trials = 100, p=.7, n=10):
    return np.random.binomial(n_trials, p,n)
```

```python
sample = simulate(n=80)
```

```python
sample
```

```python
sample.mean()
```

```python
samples_mean = [np.random.choice(sample,size=80,replace=True).mean() for _ in range(1000)]
```

```python
np.mean(samples_mean)
```

```python
np.std(samples_mean)
```

```python
plt.hist(samples_mean,bins=20) 
plt.title("histogram") 
plt.show()
```
